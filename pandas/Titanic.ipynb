### All passengers named "Miss" — what's their average age?

```python
Miss = titanic[titanic["Name"].str.contains("miss")]
Miss[["Age"]].mean(numeric_only=True).head()
```

```python
# Drop null values
titanic = titanic.dropna(subset="Embarked")
```

```python
#Fill null values
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].mean())
```

```python
titanic.isna().sum()
```

```python
# Apply method

def categorise_fare(x):
    return "expensive" if x > 100 else "cheap"
titanic["Fare"] = titanic["Fare"].apply(categorise_fare)
titanic.head()
```

```python
# Apply method with lambda
titanic["Age"] = titanic["Age"].apply(lambda x: "Adult" if x >= 18 else "Child")
titanic.head()
```

## Visualisation
### Histogram

```python
titanic = pd.read_csv("/Users/irsyad/Desktop/Python/titanic.csv")
titanic.head()
```

```python
# matplotlib version
fig,ax = plt.subplots(figsize = (8,4))
ax.hist(titanic["Age"].dropna(),bins = 20, edgecolor = 'black',color="olive")
ax.set_xlabel("Age")
ax.set_ylabel("Count")
ax.set_title("Age Distribution of Passengers")

plt.show()
```

```python
# pandas version
titanic["Age"].hist(bins = 20,edgecolor='black', figsize=(8, 4))
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Passengers")
plt.grid(visible=False)
plt.show()
```

### Barchart

```python
# Matplotlib version
survival = titanic.groupby('Pclass')["Survived"].mean()

fig, ax = plt.subplots(figsize=(8,4))
ax.bar(x=survival.index,height=survival.values)
ax.set_xlabel('Passenger Class')
ax.set_ylabel('Survival Rate')
ax.set_title('Survival Rate by Class')
plt.show()
```

```python
# Pandas version
titanic.groupby("Pclass")["Survived"].mean().plot(kind="bar",rot=0)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title("Survival Rate by Class")

plt.show
```

### Scatter Plot

```python
# Matplotlib version
fig, ax = plt.subplots(figsize=(8,4))
ax.scatter(x = titanic["Age"],y = titanic["Fare"],alpha = 0.5)
ax.set_xlabel("Age")
ax.set_ylabel("Fare")
ax.set_title("Age vs Fare")
```

```python
# Pandas version
titanic.plot.scatter(x="Age",y="Fare",alpha=0.5,figsize=(8,4))
```

### Box Plot

```python
# Matplotlib version
classes = [titanic[titanic["Pclass"]== c]["Fare"] for c in [1,2,3]]

fig,ax = plt.subplots(figsize=(8,4))
ax.boxplot(classes, labels=['Class 1', 'Class 2', 'Class 3'])
ax.set_xlabel("Passenger Classes")
ax.set_ylabel("Fare")
ax.set_title("Fare Distribution by Class")

plt.show()
```

```python
# Pandas version — much simpler!
titanic.boxplot(column='Fare', by='Pclass', figsize=(8, 4),grid = False)
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Fare Distribution by Class')
plt.show()
```

### Heatmap

```python
corr = titanic[["Age","Fare","Pclass","Survived"]].corr()
corr
```

```python
# Matplotlib version
fig,ax = plt.subplots(figsize=(8,6))
im = ax.imshow(corr,cmap="coolwarm")
ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
plt.colorbar(im)
ax.set_title("Correlation Matrix")
plt.show()
```

```python
# Seaborn version
sns.heatmap(corr, annot=True,cmap="coolwarm")
plt.title('Correlation Matrix')
plt.show()
```

```python
# 2x2 Canvas
fig,axs = plt.subplots(2,2,figsize = (14,10))

# Top left (0,0)
axs[0,0].hist(titanic["Age"].dropna(),bins=20,edgecolor="black",color="green")
axs[0, 0].set_title('Age Distribution')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Count')

# Top right (0,1)
survival = titanic.groupby('Pclass')['Survived'].mean()
axs[0, 1].bar(survival.index, survival.values, color='steelblue', edgecolor='black')
axs[0, 1].set_title('Survival Rate by Class')
axs[0, 1].set_xlabel('Passenger Class')
axs[0, 1].set_ylabel('Survival Rate')
axs[0, 1].set_xticks([1, 2, 3])

# Bottom left (1,0)
classes = [titanic[titanic['Pclass'] == c]['Fare'] for c in [1, 2, 3]]
axs[1, 0].boxplot(classes, labels=['Class 1', 'Class 2', 'Class 3'])
axs[1, 0].set_title('Fare Distribution by Class')
axs[1, 0].set_xlabel('Passenger Class')
axs[1, 0].set_ylabel('Fare')

# Bottom right (1,1)
corr = titanic[['Age', 'Fare', 'Pclass', 'Survived']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axs[1, 1])
axs[1, 1].set_title('Correlation Matrix')

plt.suptitle("Titanic EDA Dashboard', fontsize=16, fontweight='bold")
plt.tight_layout()
plt.show()
```
